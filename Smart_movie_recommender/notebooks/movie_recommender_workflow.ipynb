# Smart Movie Recommender Workflow

This notebook documents the ML workflow: data loading, preprocessing, TF-IDF feature extraction, cosine similarity, and recommendation testing.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

from src.preprocessing import save_processed_dataset
from src.recommender import MovieRecommender
```

## Step 1: Load and Explore Raw Dataset

```python
raw_movies = pd.read_csv('data/movies_raw.csv')
raw_movies.head()
```

```python
raw_movies.info()
raw_movies.isna().sum()
```

## Step 2: Preprocess and Save Cleaned Dataset

```python
processed_movies = save_processed_dataset()
processed_movies[['title', 'combined_content', 'processed_content']].head()
```

## Step 3: TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
feature_matrix = vectorizer.fit_transform(processed_movies['processed_content'])
feature_matrix.shape
```

## Step 4: Cosine Similarity Matrix

```python
cosine_sim_matrix = cosine_similarity(feature_matrix, feature_matrix)
cosine_sim_matrix.shape
```

## Step 5: Test Recommendation Function

```python
recommender = MovieRecommender()
matched_title, recommendations = recommender.get_recommendations('Inception', top_n=10)
matched_title, [movie.title for movie in recommendations]
```
